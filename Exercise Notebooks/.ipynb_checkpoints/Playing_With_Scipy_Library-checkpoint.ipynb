```python
from scipy import misc
import imageio
```

```python
f = misc.face()
```

```python
imageio.imsave('\\py practice\\Exercise Notebooks\\Data\\face_from_scipy_misc_library.png' , f)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(f) 
```

```python
f
```

```python
print(f.mean(),f.max(),f.min())
```

```python
from scipy import ndimage
```

```python
rotate_image = ndimage.rotate(f,45)
```

```python
plt.imshow(rotate_image)
plt.show()
```

```python
blur_image = ndimage.gaussian_filter(f,sigma=10)
```

```python
plt.imshow(blur_image)
plt.show()
```
